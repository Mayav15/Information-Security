## Simplified Data Encryption Standard (S-DES)
The SDES encryption algorithm takes an 8-bit block of plaintext (example: 10111101) and a 10-bit key as input and makes an 8-bit block of ciphertext as output.

```python
plain_text = "11111100"
key = "1010101101"
```

```python
class SDES:
    def __init__(self,plain_text,key):
        self.plain_text=plain_text
        self.key=key
        self.key1=""
        self.key2=""
        self.cipher_text=""
        self.plain_text_2=""
        
        # Key Generation Variables
        self.p10=[3,5,2,7,4,10,1,9,8,6]
        self.p8=[6,3,7,4,8,5,10,9] 
        
        # Encrypt and Decrypt Variables
        self.ip=[2,6,3,1,4,8,5,7]
        self.ip_1=[4,1,3,5,7,2,8,6]
        
        self.ep=[4,1,2,3,2,3,4,1]
        self.s0=[[1,0,3,2],[3,2,0,1],[0,2,1,3],[3,1,3,2]]
        self.s1=[[0,1,2,3],[2,0,1,3],[3,0,1,0],[2,1,0,3]]
        self.p4=[2,4,3,1]
        
        
    # Key Generation Functions
    def P10(self,key):
        key_ = "-"*10
        key_ = list(key_)
        for i,value in enumerate(self.p10):
            key_[i]=key[value-1]
        key_ = "".join(key_)
        return key_
    
    def P8(self,key):
        key_ = "-"*8
        key_ = list(key_)
        for i,value in enumerate(self.p8):
            key_[i]=key[value-1]
        key_ = "".join(key_)
        return key_
    
    def LS1(self,key):
        return key[1:]+key[0]
    
    def split_key(self,key):
        n=len(key)
        n=int(n/2)
        return key[:n],key[n:]
    
    def combine_key(self,key1,key2):
        return key1+key2
    
    def key_generation(self):
        key=self.key
        p10=self.P10(key)
        sk1,sk2=self.split_key(p10)
        sk1,sk2=self.LS1(sk1),self.LS1(sk2)
        key_=self.combine_key(sk1,sk2)
        key1=self.P8(key_)
        sk1,sk2=self.LS1(self.LS1(sk1)),self.LS1(self.LS1(sk2))
        key_=self.combine_key(sk1,sk2)
        key2=self.P8(key_)
        self.key1=key1
        self.key2=key2
    
    
    
    # Encrypt and Decrypt Functions
    def IP(self,text):
        text_ = "-"*8
        text_ = list(text_)
        for i,value in enumerate(self.ip):
            text_[i]=text[value-1]
        text_ = "".join(text_)
        return text_
    
    def IP_1(self,text):
        text_ = "-"*8
        text_ = list(text_)
        for i,value in enumerate(self.ip_1):
            text_[i]=text[value-1]
        text_ = "".join(text_)
        return text_
    
    def EP(self,text):
        text_ = "-"*8
        text_ = list(text_)
        for i,value in enumerate(self.ep):
            text_[i]=text[value-1]
        text_ = "".join(text_)
        return text_
    
    def XOR(self,text1,text2):
        xor = ''
        for i in range(len(text2)):
            xor+='0' if text1[i]==text2[i] else '1'
        return xor
    
    def S0(self,text):
        row = int(str(int(text[0])*10+int(text[3])),2)
        column = int(str(int(text[1])*10+int(text[2])),2)
        val = self.s0[row][column]
        val = str(bin(val).replace("0b", ""))
        if val=='0':
            val='00'
        elif val=='1':
            val="01"
        return val
    
    def S1(self,text):
        row = int(str(int(text[0])*10+int(text[3])),2)
        column = int(str(int(text[1])*10+int(text[2])),2)
        val = self.s1[row][column]
        val = str(bin(val).replace("0b", ""))
        if val=='0':
            val='00'
        elif val=='1':
            val="01"
        return val
    
    def P4(self,text):
        text_ = "-"*4
        text_ = list(text_)
        for i,value in enumerate(self.p4):
            text_[i]=text[value-1]
        text_ = "".join(text_)
        return text_
    
    def FX(self,text,key):
        text1,text2=self.split_text(text)
        e_text2=self.EP(text2)
        text=self.XOR(e_text2,key)
        t1,t2=self.split_text(text)
        t1,t2=self.S0(t1),self.S1(t2)
        text=self.merge_text(t1,t2)
        text=self.P4(text)
        text1=self.XOR(text1,text)
        text=self.merge_text(text1,text2)
        return text
        
    def split_text(self,text):
        n=len(text)
        n=int(n/2)
        return text[:n],text[n:]
    
    def merge_text(self,text1,text2):
        return text1+text2
    
    def swap_text(self,text1,text2):
        return text2,text1
    
    def encrypt_text(self):
        key1 = self.key1
        key2 = self.key2
        plain_text = self.plain_text
        text = self.IP(plain_text)
        text = self.FX(text,key1)
        text1,text2 = self.split_text(text)
        text1,text2 = self.swap_text(text1,text2)
        text = self.merge_text(text1,text2)
        text = self.FX(text,key2)
        cipher_text = self.IP_1(text)
        self.cipher_text=cipher_text
    
    def decrypt_text(self):
        key1 = self.key1
        key2 = self.key2
        cipher_text = self.cipher_text
        text = self.IP(cipher_text)
        text = self.FX(text,key2)
        text1,text2 = self.split_text(text)
        text1,text2 = self.swap_text(text1,text2)
        text = self.merge_text(text1,text2)
        text = self.FX(text,key1)
        plain_text = self.IP_1(text)
        self.plain_text_2=plain_text
    
    # final function to execute everything
    def execute_all(self):
        print("Input")
        print("Key (10 bit):\t\t",self.key)
        print("Plain Text (8 bit):\t",self.plain_text)
        print("\n")
        print("Output")
        self.key_generation()
        print("Key 1 (8 bit):\t\t",self.key1)
        print("Key 2 (8 bit):\t\t",self.key2)
        print("\n")
        self.encrypt_text()
        print("After Encryption")
        print("Cipher Text (8 bit):\t",self.cipher_text)
        print("\n")
        S_DES.decrypt_text()
        print("After Decryption")
        print("Plain Text (8 bit):\t",self.plain_text_2)
        print("\n")
```
